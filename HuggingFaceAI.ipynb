```python
from transformers import pipeline
```

```python
sentiment_analyzer = pipeline("sentiment-analysis")

result = sentiment_analyzer("I love studying Artificial Intelligence!")
result
```

```python
classifier = pipeline("text-classification")

classifier("This workshop on Generative AI is very useful for students")
```

```python
translator = pipeline("translation_en_to_fr")

translator("Artificial Intelligence is transforming education.")
```

```python
summarizer = pipeline("summarization")

text = """
Generative AI refers to artificial intelligence systems that are capable of
generating text, images, audio, and video. These systems are widely used in
education, healthcare, and industry.
"""

summarizer(text, max_length=50, min_length=20, do_sample=False)
```

```python
embedding_model = pipeline("feature-extraction")

embeddings = embedding_model("AI is the future of technology")

len(embeddings[0][0])
```
